```python
Scores = {'James':82, 'Mary':98, 'Yamamoto':93}
# key1:value1, key2:value2, ..., key_n:value_n
# Key: string, comparable >, <, ==,
# value: anything
print(type(Scores))
print(Scores['Mary'])	# 98
Scores['Yamamoto'] += 7
print(Scores)		# {'James':82, 'Mary':98, 'Yamamoto':100}
print(len(Scores))	# 3
```

```python
Scores = {'James':(82, 100, 85),
          'Mary':(98, 98, 98),
          'Yamamoto':(93, 80, 82)}
# key1:value1, key2:value2, ..., key_n:value_n
# Key: string, comparable >, <, ==,
# value: anything
print(type(Scores))
print(Scores['James'])	# 98
Scores['Yamamoto'] += (7,)
print(Scores)		# {'James':82, 'Mary':98, 'Yamamoto':100}
print(len(Scores))	# 3
```

```python
Scores = {'James':82,
          'Mary':98,
          'Yamamoto':93,
          'James':90} # updating, replacing
print(Scores)
print(Scores['James'])
Scores['Andy'] = 85
print(Scores)
Scores.update({'Bob':65, 'Judy':100})
print(Scores)
print(len(Scores))
print()
for x in Scores:
  print(x, Scores[x])

print()
K = Scores.keys()
print(K, type(K))

V = Scores.values()
print(V, type(V))

T = Scores.items()
print(T, type(T))

print()
for t in Scores.items():
  print(t, type(t), len(t))
  print(t[0], t[1])
```

```python
from collections import Counter
s = 'cccbbbaaabbbaaabbb'
cnt = Counter(s)
print('----------')
print(type(cnt))
print(cnt)		# Counter({'b': 9, 'a': 6, 'c': 3})
print('----------')
print(cnt.keys())	# dict_keys(['c', 'b', 'a']) print('----------')
print(cnt.items())	# dict_items([('c', 3), ('b', 9), ('a', 6)])
```
