```python
L1 = [0, 1, 2] # memory address
L2 = L1  # memory address
print(L1)
print(L2)
L1[0] += 10
print(L1)
print(L2)
```

```python
L1 = [0, 1, 2] # memory address of original data
L2 = list(L1)  # memory address of new data
# Shallow copy; only copy the elements of highest dimension
print(L1)
print(L2)
L1[0] += 10
print(L1)
print(L2)
```

```python
x = int(input('Input an integer:'))
if x % 2 == 0:
    print(x, 'is even')
    print('.')
else:
    print(x, 'is odd')
print('!')
```

```python
score = int(input('Input a score:'))
if score >= 60:
    print('Pass')
    if score >= 90:
      print('Execellent')
      print('@')
    elif score >= 80: # else if
      print('Good')
      print('#')
else:
    print('Failed')
print('!')
```
