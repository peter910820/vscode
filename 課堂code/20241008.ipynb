```python
# Ch 1
# 1-1
# 1-1-1
# 1-1-2
# 1-1-3
# 1-2
# 1-3
# 1-4
# 1-5
# Ch 2
# 2-1
# 2-2
# ...

ch = 1
while ch <= 5:
  print('Ch', ch)
  section = 1
  while section <= 6:       ##
    print(ch, '-', section) ##
    subsection = 1
    while subsection <= 3:  ##
      print(ch, '-', section, '-', subsection)
      subsection += 1       ##
    section += 1            ## 2nd layer while statement
  ch += 1

```

```python
n = 0
sum = 0
while n >= 0:
    n = int(input('Input a non-negative integer: '))
    if n >= 0:
        sum += n
print('sum is', sum)
```

```python
s = ''
sOut = ''
while s != '.':
    s = input('Input a word: ')
    if s == '.':
        sOut = sOut + s
    else:
        sOut = sOut + ' ' + s
print(sOut)
```
